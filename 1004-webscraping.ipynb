```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import requests
website_url = requests.get('https://en.wikipedia.org/wiki/2019_ATP_Tour').text
```

```python
from bs4 import BeautifulSoup
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
table = soup.find_all('table',{'class':'wikitable'})[15]
table
```

```python
trs = table.find_all('tr')[1:]
trs
```

```python
trs
```

```python
table = []

colnames = []
for th in trs[0].find_all('th'):
    colnames.append(th.getText()[:-1])

for tr in trs[1:]:
    row = []       
    for td in tr.find_all('td'):
        row.append(td.getText())
        
    table.append(row)

for i in range(len(table)):
    table[i][1] = table[i][1].replace('\xa0','')
    table[i][2] = table[i][2].replace(',','')
    table[i][4] = table[i][4].replace('\n','')

table
```

```python
data = pd.DataFrame(table,columns=colnames)
data.set_index('#',inplace=True)
data.drop('Move†',axis=1,inplace=True)
data
```

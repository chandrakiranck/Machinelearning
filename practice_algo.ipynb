```
X=data.drop(["Target"],axis=1)
Y=data.Target
```

```
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
for col in data.columns:
  data[col]=le.fit_transform(data[col])
```

```
from sklearn.preprocessing import MinMaxScaler

```

```
mm=MinMaxScaler()
Xs=mm.fit_transform(X)
```

```
 knn = KNeighborsClassifier()
dt=DecisionTreeClassifier(criterion="entropy",random_state=0)
KNN_final=KNeighborsClassifier(n_neighbors=452,weights="distance")
#KNN_2=KNeighborsClassifier(n_neighbors=14,weights="distance")
lr=LogisticRegression()
rf=RandomForestClassifier(n_estimators=10,random_state=0)
KNN_bag=BaggingClassifier(base_estimator=knn,n_estimators=10,random_state=0)
lr_bag=BaggingClassifier(base_estimator=lr,n_estimators=10,random_state=0)
nb_bag=BaggingClassifier(base_estimator=NB,n_estimators=10,random_state=0)
dt_bag=BaggingClassifier(n_estimators=10,random_state=0)

#BOOSTING MODELS
lr_boost=AdaBoostClassifier(base_estimator=lr,n_estimators=50,random_state=0)
nb_boost=AdaBoostClassifier(base_estimator=NB,n_estimators=50,random_state=0)
dt_boost=AdaBoostClassifier(n_estimators=50,random_state=0)

#gradienttboosting

gboost=GradientBoostingClassifier(n_estimators=300,random_state=0)
```

```
NB=GaussianNB()
lr=LogisticRegression()
models=[]
models.append(("KNN_WV",KNN_final))
models.append(("NB",NB))
models.append(("Logistic",lr))
models.append(("fgdt",dt))
models.append(("random_frst",rf))
models.append(("KNN_bag",KNN_bag))
models.append(("NB_bag",nb_bag))
models.append(("Logisticbag",lr_bag))
models.append(("dtbag",dt_bag))
models.append(("NB_boost",nb_boost))
models.append(("Logisticboost",lr_boost))
models.append(("dtboost",dt_boost))
models.append(("gradientboost",gboost))
```

```
from sklearn.model_selection import KFold, cross_val_score
results=[]
names=[]
for name,model in models:
    kfold=KFold(shuffle=True,n_splits=15,random_state=0)
    cv_results=cross_val_score(model,Xs,Y,cv=kfold,scoring='f1_weighted')
    results.append(cv_results)
    names.append(name)
    print("%s %f (%f)"% (name,np.mean(cv_results),np.var(cv_results,ddof=1)))
#boxplot algorithm
fig=plt.figure()
fig.suptitle("Algorithm Comparsion")
ax=fig.add_subplot(111)
plt.boxplot(results)
ax.set_xticklabels(names)
plt.show()
```
